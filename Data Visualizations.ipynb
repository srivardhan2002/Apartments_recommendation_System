# Data Visualization

This notebook will focus on data visualizations of the dataset before proceeding with the main algorithm of recommendation system.

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the apartments dataset
apartments_df = pd.read_csv('Datasets/Cleaned_House_Rent_Dataset.csv')
apartments_df.head()
```

### Analysis of Rents

```python
rents = apartments_df['Rent'].to_numpy()
rents.shape
```

```python
plt.hist(rents,  bins = 100, edgecolor = 'black', alpha = 0.7)
# plt.xticks(range(10))
plt.grid(True)
plt.show()
```

```python
#considering only the rents of flats with are below 50k.
rents_low = rents[rents <= 5e4]
rents_low.shape
```

```python
# plotting histogram for rents
plt.hist(rents_low,  bins = 100, edgecolor = 'black', alpha = 0.7)
# plt.xticks(range(10))
plt.grid(True)
plt.xlabel("Rents of flats")
plt.ylabel("Frequency")
plt.show()
```

```python
print('Distribution of Rents')
for i in range(10):
    print(i*5000, '-', (i+1)*5000, ':', rents[(rents >= i*5000) & (rents < (i+1)*5000)].shape[0])

print("above 50k:", rents[rents >= 50000].shape[0])
```

### Analysis of Sizes

```python
sizes = apartments_df['Size'].to_numpy()
sizes.sort()
sizes
```

```python
plt.hist(sizes, bins = 100, edgecolor = 'black', alpha = 0.7)
plt.xlabel('Sizes of flats')
plt.ylabel('Frequeny')
plt.grid(True)
plt.show()
```

```python
print('Distribution of Sizes')
for i in range(6):
    print(i*500, '-', (i+1)*500, ":", sizes[(sizes >= i*500)&(sizes < (i+1)*500)].shape[0])

print("abvoe 3k :", sizes[sizes >= 3000].shape[0])
```

### Scatter plots

```python
#considering only the flats whose rent is less than 50k and size less than 3k.
tmp_df = apartments_df[(apartments_df['Rent'] < 50000) & (apartments_df['Size'] < 3000) ]
tmp_df.shape
```

```python
plt.scatter(tmp_df['Size'], tmp_df['Rent'],  color='blue', label='Sample Data')
plt.xlabel('Sizes')
plt.ylabel('Rents')
plt.title('Sample Scatter Plot')

plt.show()
```

```python
plt.scatter(tmp_df['BHK'], tmp_df['Rent'],  color='blue', label='Sample Data')
plt.xlabel('No of bedrooms')
plt.ylabel('Rent')
plt.title('Sample Scatter Plot')

plt.show()
```
